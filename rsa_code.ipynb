```python
import math
import random
from sympy import isprime
```

```python
def is_prime(n, k=20):
    if n <= 1:
        return False
    if n <= 3:
        return True
    if n % 2 == 0:
        return False

    s, d = 0, n - 1
    while d % 2 == 0:
        s += 1
        d //= 2

    for _ in range(k):
        a = random.randint(2, n - 2)
        x = pow(a, d, n)
        if x == 1 or x == n - 1:
            continue

        for _ in range(s - 1):
            x = pow(x, 2, n)
            if x == n - 1:
                break
        else:
            return False
    return True
```

```python
def is_relative_prime(a, b):
    return math.gcd(a, b) == 1
```

```python
def square_and_multiply(baza, eksp, mod):
  #želimo izračunati x^K mod N
  # x je baza, K je eksponent, N je mod
  r=1
  for bit in bin(eksp)[2:]:
    r= (r * r) % mod
    if bit == '1':
      r=(r * baza) % mod
  return r


x = 5
e = 13
modn = 19

rezultat = square_and_multiply(x,e,modn)
print(f"Rezultat {x}^{e} mod {modn} je {rezultat}")
```

```python
def montgomery_ladder(x, e, modn):
    R0 = 1
    R1 = x

    for i in range(e.bit_length() - 1, -1, -1):
        if (e >> i) & 1:
            R0 = (R0 * R1) % modn
            R1 = (R1 * R1) % modn
        else:
            R1 = (R0 * R1) % modn
            R0 = (R0 * R0) % modn

    return R0

x = 5
e = 13
modn = 19

rezultat = montgomery_ladder(x, e, modn)
print(f"Rezultat {x}^{e} mod {modn} je {rezultat}")
```

```python
def modular_exponentiation(base, exp, mod):
    result = 1
    base = base % mod
    while exp > 0:
        if exp % 2 == 1:
            result = (result * base) % mod
        base = (base * base) % mod
        exp //= 2
    return result
```

```python
from math import sqrt

def multiplicative_splitting(x, N):
    a0, a1 = 0, 1
    r0, r1 = N, x
    c = int(sqrt(N)) + 1

    while abs(r1) >= c:
        q = r0 // r1
        r0, r1 = r1, r0 - q * r1
        a0, a1 = a1, a0 - q * a1

    x0 = a1
    x1 = r1
    return x0, x1
```

```python
def modular_exponentiation_with_splitting(x, e, N):

    x0, x1 = multiplicative_splitting(x, N)

    # inverz
    x0_inv = pow(x0, -1, N)

    part1 = modular_exponentiation(x0_inv, e, N)
    part2 = modular_exponentiation(x1, e, N)

    result = (part1 * part2) % N
    return result


#test
x=5
e=13
N=19
result = modular_exponentiation_with_splitting(x, e, N)
print(f"{x}^{e} mod {N} = {result}")
```

```python
def encrypt(message, e, n, algorithm):
    if algorithm == 'Square-and-Multiply':
        return square_and_multiply(message, e, n)
    elif algorithm == 'Montgomery Ladder':
        return montgomery_ladder(message, e, n)
    elif algorithm == 'Multiplicative Splitting modulo n':
        return modular_exponentiation_with_splitting(message, e, n)
    else:
        raise ValueError("Nepoznat algoritam!")
```

```python
def get_positive_integer(prompt):
    while True:
        try:
            value = int(input(prompt))
            if value <= 0:
                print("Unesite broj veći od nule.")
            else:
                return value
        except ValueError:
            print("Neispravan unos. Molimo unesite cijeli broj.")
```

```python
def check_prime_pair(p, q):
    if p == q:
        print("Brojevi p i q ne mogu biti isti. Pokušajte ponovno.")
        return False
    if not is_prime(p) or not is_prime(q):
        print("Pogrešan unos. p i q moraju biti prosti brojevi.")
        return False
    return True
```

```python
def calculate_private_key(e, phi):
    try:
        return pow(e, -1, phi)
    except ValueError:
        print("Ne može se izračunati privatni ključ. Provjerite unos eksponenta e i ϕ(n).")
        exit()
```

```python
def get_message_input(n):
    while True:
        try:
            message = int(input(f"Unesite poruku (broj manji od n={n}): "))
            if message >= n:
                print(f"Poruka mora biti manja od n ({n}). Pokušajte ponovno.")
            elif message <=0:
              print("Poruka mora biti veća od 0")
            else:
                return message
        except ValueError:
            print("Pogrešan unos. Molimo unesite broj.")
```

```python
def decrypt(ciphertext, private_key, algorithm):
    try:
        d,n = private_key
        if not (0 <= ciphertext < n):
            raise ValueError("Ciphertext must be in the range [0, n).")

        if algorithm =='Square-and-Multiply':
            return square_and_multiply(ciphertext, d,n)
        elif algorithm == 'Montgomery Ladder':
            return montgomery_ladder(ciphertext,d,n)
        elif algorithm == 'Multiplicative Splitting modulo n':
            return modular_exponentiation_with_splitting(ciphertext, d,n)
        else:
            raise ValueError("Invalid decryption algorithm selected.")
    except Exception as ex:
        print(f"Error during decryption: {ex}")
        raise
```

```python
def main():

    while True:
        choice = input("Želite li sami unijeti proste brojeve (1) ili da ih program generira (2)? (1/2): ")
        if choice == '1' or choice == '2':
            break
        else:
            print("Nevažeći unos. Molimo unesite '1' za unos vlastitih brojeva ili '2' za generiranje random brojeva.")

    if choice == '1':
        while True:
            p = get_positive_integer("Unesite prost broj p: ")
            if is_prime(p):
                break
            print("Broj nije prost. Pokušajte ponovno.")

        while True:
            q = get_positive_integer("Unesite prost broj q: ")
            if is_prime(q):
                if check_prime_pair(p,q):
                  break


            else:
              print("Broj nije prost. Pokušajte ponovno.")


    elif choice == '2':
        min_val = 2
        max_val = 100000000
        print(f"Generiramo slučajne proste brojeve između {min_val} i {max_val}.")
        p = generate_random_prime(min_val, max_val)
        q = generate_random_prime(min_val, max_val)
        print(f"Generirani prosti brojevi: p = {p}, q = {q}")

    n = p * q
    phi = (p - 1) * (q - 1)


    while True:
        print(f"ϕ(n) je {phi}")
        e = get_positive_integer("Unesite javni eksponent e (relativno prost s ϕ(n)): ")
        if is_relative_prime(e, phi):
            break
        print("e nije relativno prost s ϕ(n). Pokušajte ponovno.")

    d = calculate_private_key(e, phi)

    print(f"Javni ključ: (n={n}, e={e})")
    print(f"Privatni ključ: (n={n}, d={d})")

    message = get_message_input(n)

    print("Odaberite algoritam:")
    print("1. Square-and-Multiply")
    print("2. Montgomery Ladder")
    print("3. Multiplicative Splitting modulo N")

    while True:
        choice = input("Unesite broj algoritma (1/2/3): ")
        if choice in ['1', '2', '3']:
            break
        print("Nevažeći unos. Molimo unesite broj između 1 i 3.")

    algorithms = {
        '1': 'Square-and-Multiply',
        '2': 'Montgomery Ladder',
        '3': 'Multiplicative Splitting modulo n'
    }
    algorithm = algorithms[choice]


    ciphertext = encrypt(message, e, n, algorithm)
    print(f"Šifrirana poruka: {ciphertext}")

    print("Želite li dešifrirati poruku? Za DA unesite 1, a za izlaz unesite bilo što!")

    while True:
      izbor = input()
      if izbor == '1':
        decrypted_message = decrypt(ciphertext, (d,n), algorithm)
        print("Decrypted message:", decrypted_message)
        break
      else:
        print("Odustali ste :) ")
        return

```
